```python
print("calculadora simple")
print(1, "suma")
print(2, "resta")
print(3, "multiplicacion")
print(4, "dividir")
opcion=float(input("seleccione una operacion(1_4):"))
number1 = float(input("ingrese numero 1:"))
if opcion==1 or opcion=="+":
    number2 = float(input("ingrese numero 2:"))
    print("vamos a sumar(+) los dos valores ingresados:")
    print('La suma es:',number1+number2)
if opcion==2 or opcion=="-":
    number2 = float(input("ingrese numero 2: "))
    print("vamos a restar(-) los dos valores ingresados:\n")
    print('La resta es:',number1-number2)
if opcion==3 or opcion=="*":
    number2 = float(input("ingres e numero 2: "))
    print("vamos a multiplicar(*) los dos valores ingresados:\n")
    print('La multiplicacion es:',number1*number2)
if opcion==4 or opcion=="/":
    number2 = float(input("ingrese numero 2: "))
    print("vamos a dividir(/) los dos valores ingresados:\n")
    print('La divicion es:',number1/number2)
```
